# Liste

In questa lezione vedremo uno dei più utili tipi predefiniti di Python: le liste. Come una stringa, una **lista** è una sequenza di valori. Mentre in una stringa i valori sono dei caratteri, in una lista possono essere di qualsiasi tipo. I valori che fanno parte della lista sono chiamati **elementi**.

Liste.png

Ci sono vari modi per creare una nuova lista; quello più semplice consiste nel racchiudere i suoi elementi tra parentesi quadrate (`[` e `]`).
Nella cella qui sotto, il primo esempio è una lista di quattro interi; il secondo è una lista di tre stringhe. 

Non è necessario che gli elementi di una stessa lista siano tutti dello stesso tipo: il terzo esempio è una lista che contiene una stringa, un numero in virgola mobile, un intero e (meraviglia!) un'altra lista. Una lista all'interno di un'altra lista è detta lista **nidificata**.

Una lista che non contiene elementi è detta **lista vuota**; potete crearne una scrivendo le due parentesi quadre vuote, `[]`.

````
numeri = [10, 20, 30, 40]
menu = ['Primi piatti', 'Secondi piatti', 'Dessert']
varie = ['spam', 2.0, 5, [10, 20]]
vuota = []
print(numeri, menu, varie, vuota)
````

## Le liste sono mutabili

La sintassi per accedere agli elementi di una lista è la stessa usata per i caratteri di una stringa: le parentesi quadre, con un'espressione tra parentesi che specifica l’indice dell’elemento. 

Ricordate che gli indici partono da 0!

````
menu[0]
````

A differenza delle stringhe, le liste sono *mutabili*. Quando l’operatore parentesi quadre compare sul lato sinistro di un’assegnazione, identifica l’elemento della lista che sarà riassegnato:

````
numeri = [10, 20, 30]
numeri[1] = 5
print(numeri)
````

Gli indici delle liste funzionano nello stesso modo di quelli delle stringhe:
* L'indice può essere qualsiasi espressione di tipo intero.
* Se tentate di leggere o modificare un elemento che non esiste, ottenete un errore `IndexError`.
* Con un indice di valore negativo, si conta a ritroso dalla fine della lista.

## Attraversamento di una lista

Parecchi tipi di calcolo comportano l'elaborazione di una lista un elemento per volta. Spesso iniziano dal primo elemento, selezionano un elemento per volta, eseguono una certa operazione e continuano fino alla fine della lista. Questo tipo di elaborazione è detta **attraversamento**. Un modo per scrivere un attraversamento è quello di usare un ciclo `while`. Per sapere quanti sono gli elementi presenti in una lista possiamo usare la funzione predefinta `len`.


````
menu = ['Primi piatti', 'Secondi piatti', 'Dessert']

i = 0
while i < len(menu):
  piatto = menu[i]
  print(piatto)
  i = i + 1
````

Un altro modo di scrivere un attraversamento è usare un ciclo `for`. 
Nell'esempio qui sotto, ad ogni ciclo il successivo elemento della lista viene assegnato alla variabile `piatto`. Il ciclo continua finché non rimangono più caratteri da analizzare.

````
menu = [1, 2, 3]

for p in menu:
  print(p)
````

Questo metodo funziona bene per leggere gli elementi di una lista, ma se volete scrivere o aggiornare degli elementi vi servono gli indici. Un modo per farlo è usare una combinazione delle funzioni predefinite `range` e `len`.

Questo esempio attraversa la lista e aggiorna tutti gli elementi. `len` restituisce il numero di elementi della lista. `range` restituisce una lista di indici da `0` a `n − 1`, dove `n` è la lunghezza della lista. Ad ogni ripetizione del ciclo, `i` prende l'indice dell'elemento successivo. L'istruzione di assegnazione nel corpo usa `i` per leggere il vecchio valore dell'elemento e assegnare quello nuovo.

````
numeri = [10, 20, 30]

for i in range(len(numeri)):
    numeri[i] = numeri[i] * 2

print(numeri)
````

## Operazioni sulle liste

In Pyhton l'operatore `+` concatena delle liste, mentre l'operatore `*` ripete una lista per un dato numero di volte:

````
a = [1, 2, 3]
b = [4, 5, 6]
c = a + b
print(c)
d = a * 3
print(d)
````

Gli operatori `+` e `*` si possono usare anche con le strighe, con lo stesso significato:

````
a = 'bagno'
b = 'schiuma'
c = a + b
print(c)
d = c * 3
print(d)
````

## Metodi delle liste

Python fornisce dei metodi che operano sulle liste. Ad esempio, `append` aggiunge un nuovo elemento in coda alla lista:

````
t = ['a', 'b', 'c']
t.append('d')
print(t)
````

`extend` prende una lista come argomento e accoda tutti i suoi elementi. Questo esempio lascia immutata la lista `t2`.

````
t1 = ['a', 'b', 'c']
t2 = ['d', 'e']
t1.extend(t2)
print(t1)
t3 = t1 + t2
print(t1)
print(t2)
print(t3)
````

`sort` dispone gli elementi della lista in ordine crescente:

````
t = ['d', 'c', 'e', 'b', 'a']
t.sort()
print(t)
````

## Cancellare elementi

Ci sono vari modi per cancellare elementi da una lista. Se conoscete l’indice dell’elemento desiderato, potete usare `pop`:

````
t = ['a', 'b', 'c']
x = t.pop(1)
print(t)
print(x)
````

`pop` modifica la lista e restituisce l’elemento che è stato rimosso. Se omettete l’indice, il metodo cancella e restituisce l’ultimo elemento della lista. 

Se non vi serve il valore rimosso, potete usare l’operatore `del`:

````
t = ['a', 'b', 'c']
del t[1]
print(t)
````

Se conoscete l’elemento da rimuovere ma non il suo indice, potete usare `remove`:

````
t = ['a', 'b', 'c']
t.remove('b')
print(t)
````

## Oggetti, valori e alias

Se eseguiamo queste istruzioni di assegnazione:

```python
a = 'banana'
b = 'banana'
```

Sappiamo che `a` e `b` si riferiscono a una stringa, ma non sappiamo se si riferiscono alla *stessa stringa*. Ci sono due possibili stati della memoria dell'interprete Python:

* In un caso, `a` e `b` si riferiscono a due oggetti diversi che hanno lo stesso valore. 
* Nell’altro, si riferiscono allo stesso oggetto.

image_2022-10-26_190435921.png

Per controllare se due variabili si riferiscono allo stesso oggetto, potete usare l’operatore `is`.

````
a = 'banana'
b = 'banana'

a is b
````

Ma se create due liste, ottenete due oggetti distinti:

````
a = [1, 2, 3]
b = [1, 2, 3]

print(a is b)

a[0] = 0
b[2] = 0
print(a)
print(b)
````

In quest’ultimo caso si dice che le due liste sono **equivalenti**, perché contengono gli stessi elementi, ma non **identiche**, perché non sono lo stesso oggetto. Se due oggetti sono identici, sono anche equivalenti, ma se sono equivalenti non sono necessariamente identici.

Se la variabile `a` si riferisce a un oggetto e assegnate `b = a`, allora entrambe le variabili si riferiscono allo stesso oggetto.

````
a = [1, 2, 3]
b = a

b is a

a[0] = 0
print(a)
print(b)
````

L’associazione tra una variabile e un oggetto è chiamato **riferimento**. Nell'esempio precedente ci sono due riferimenti allo stesso oggetto.
Un oggetto che ha più di un riferimento ha anche più di un nome, e si dice quindi che l’oggetto ha degli **alias**.

Se l’oggetto munito di alias è mutabile, i cambiamenti provocati da un alias si riflettono anche sull’altro:

````
b[0] = 42
print(a)
````

Sebbene questo comportamento possa essere utile, è anche fonte di errori. In genere è più sicuro *evitare gli alias quando si sta lavorando con oggetti mutabili*.

## Liste come argomenti

Quando passate una lista a una funzione, questa riceve un riferimento alla lista. Se la funzione modifica la lista, il chiamante vede la modifica. Per esempio, `decapita` rimuove il primo elemento di una lista:

````
def decapita(t):
    del t[0]

lettere = ['a', 'b', 'c']
decapita(lettere)
print(lettere)
````

Il parametro `t` e la variabile `lettere` sono due alias dello stesso oggetto.

È importante distinguere tra operazioni che *modificano* le liste e operazioni che *creano nuove liste*. Per esempio il metodo `append` modifica una lista, ma l’operatore `+` ne crea una nuova.

Questa differenza è importante quando scrivete delle funzioni che devono modificare delle liste. Un’alternativa valida è scrivere una funzione che *crea e restituisce* una nuova lista. Per esempio, `ritaglia` restituisce tutti gli elementi di una lista tranne il primo, lasciando intatta la lista di origine:

````
def ritaglia(t):
  t = t.copy()    # crea una copia di t
  decapita(t)       
  return t

lettere = ['a', 'b', 'c']
resto = ritaglia(lettere)
print(resto)
print(lettere)
````

## Esercizio 1
Scrivere una funzione ``somma`` che ritorna la somma di tutti gli elementi di una lista numerica `L` passata come parametro. Risolvete l'esercizio prima di guardare la soluzione.

### Soluzione esercizio 1 liste

````
def somma(L):
    totale = 0
    for x in L:
        totale = totale + x
    return totale

numeri = [10, 30, 20]
somma(numeri)
````

`totale` è inizializzato a `0`. Ad ogni ripetizione del ciclo,
`x` prende un elemento dalla lista. Man mano che il ciclo lavora, `totale` accumula la somma degli elementi; una variabile usata in questo modo è detta anche `accumulatore`.

Sommare gli elementi di una lista è un'operazione talmente comune che Python contiene un'apposita funzione predefinita, `sum`:

````
L = [1, 2, 3]
sum(L)
````

## Esercizio 3
Scrivere una funzione ``massimo`` che ritorna il massimo di una lista numerica `L` passata come parametro. Invocare la funzione su una lista di prova. Risolvete l'esercizio prima di guardare la soluzione.


### Soluzione esercizio 3 Liste

````
def massimo(L):
  m = L[0]
  for x in L:
    if x > m:
      m = x
  return m

numeri = [10, 30, 20]
massimo(numeri)
````

`m` è inizializzato al primo elemento della lista. Ad ogni ripetizione del ciclo, `x` prende un elemento dalla lista. Se il valore di `x` è maggiore del valore corrente di `m`, il nuovo valore di `m` diventa `x`. In questo modo, man mano che il ciclo lavora, `m` contiene il valore più grande che è stato visto fino a quel momento. Alla fine del ciclo, `m` conterrà il massimo della lista.

Python contiene le funzioni predefinite, `min` e `max` per trovare il minimo ed il massimo di una sequenza:

````
L = [2, 1, 4, 3, 6]
print(min(L))
print(max(L))
````
